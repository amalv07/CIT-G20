```python
import pandas as pd
comparison_pivot = pd.read_csv("data.csv")
```

```python
# Bar Chart for YOLO model comparison
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
comparison_pivot.set_index("Metric").plot(kind="bar", figsize=(10, 6))

plt.xlabel("Metrics")
plt.ylabel("Values")
plt.title("Bar Chart Comparison of YOLO Models Across Metrics")
plt.legend(title="YOLO Models")
plt.grid(axis="y")
plt.xticks(rotation=45)
plt.show()

# Scatter Plot for YOLO model comparison
plt.figure(figsize=(10, 6))
for model in comparison_pivot.columns[1:]:
    plt.scatter(comparison_pivot["Metric"], comparison_pivot[model], label=model, s=100)

plt.xlabel("Metrics")
plt.ylabel("Values")
plt.title("Scatter Plot Comparison of YOLO Models Across Metrics")
plt.legend(title="YOLO Models")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Pie Chart for one metric (mAP50) across YOLO models
plt.figure(figsize=(8, 8))
plt.pie(comparison_pivot.loc[comparison_pivot["Metric"] == "mAP50", ["YOLOv8", "YOLOv9", "YOLOv10"]].values.flatten(),
        labels=["YOLOv8", "YOLOv9", "YOLOv10"], autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart of mAP50 across YOLO Models")
plt.show()
```
